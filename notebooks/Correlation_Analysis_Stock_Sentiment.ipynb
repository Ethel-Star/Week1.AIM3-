# Correlation Analysis 

Some stocks' sentiment score distributions, such as those of NVDA and TSLA, show an essentially symmetrical form with the neutral threshold at its center. This balance, especially when the data is aggregated over multiple days, suggests a distribution that is rather close to normal. In terms of other equities, like GOOG, AMZN, and AAPL, the overall trend points to a distribution that is comparable to normal, even though there may be some skewness or outliers.

It makes sense to look into if there is a linear relationship between changes in sentiment scores and changes in stock returns given this approximate normal distribution. Since the Pearson correlation assesses both the strength and direction of the linear relationship between two variables, it is a useful tool for this kind of work.

The most effective use of Pearson correlation is when the variables follow a normal distribution, a common assumption for daily returns and average sentiment scores, particularly once they have been aggregated. Hence, it is suitable to utilize Pearson correlation to examine the connection between sentiment and stock returns in this scenario.

# Aggregate Sentiments and Correlation Analysis :
For further analysis let us compute average daily sentiment scores for multiple articles that appears on the same day. 

```python
output_directory = r'C:\Users\ethio\Desktop\Tenx.AIM2 -ET\Week1\Data\Avg_Senti'
os.makedirs(output_directory, exist_ok=True)
for stock, path in aligned_data_paths.items():
    df = pd.read_csv(path)  
    merged_df = fa.compute_and_merge_sentiment(df) 
    output_filename = f'{stock.lower()}_avg_sentiment.csv'
    output_path = os.path.join(output_directory, output_filename)
    merged_df.to_csv(output_path, index=False)
   # print(f"{stock}: Merged data with average daily sentiment saved to {output_path}")
```

Then we calculate the Pearson correlation coefficient between these average daily sentiment scores and the corresponding daily returns for each stock. The Pearson correlation coefficient measures the strength and direction of the linear relationship between two variables.

```python
avg_sentiment_paths = {
    'AAPL': r'C:\Users\ethio\Desktop\Tenx.AIM2 -ET\Week1\Data\Avg_Senti\aapl_avg_sentiment.csv',
    'AMZN': r'C:\Users\ethio\Desktop\Tenx.AIM2 -ET\Week1\Data\Avg_Senti\amzn_avg_sentiment.csv',
    'GOOG': r'C:\Users\ethio\Desktop\Tenx.AIM2 -ET\Week1\Data\Avg_Senti\goog_avg_sentiment.csv',
    'NVDA': r'C:\Users\ethio\Desktop\Tenx.AIM2 -ET\Week1\Data\Avg_Senti\nvda_avg_sentiment.csv',
    'TSLA': r'C:\Users\ethio\Desktop\Tenx.AIM2 -ET\Week1\Data\Avg_Senti\tsla_avg_sentiment.csv'
}
start_date_filtered = pd.to_datetime('2020-04-27')
end_date_filtered = pd.to_datetime('2020-06-10')
print("Filtered Date Results:")
fa.plot_and_calculate_correlations_avg(avg_sentiment_paths, filter_dates=(start_date_filtered, end_date_filtered))

# Plot and calculate correlations for unfiltered data using date range from the data
print("Unfiltered Date Results:")
fa.plot_and_calculate_correlations_avg(avg_sentiment_paths)
```

The relationship between average daily sentiment scores and daily returns is extremely poor for all of the equities under analysis. This implies that there is little correlation between sentiment scores obtained from news stories and daily stock performance. Sentiment scores by themselves are an unreliable predictor for anticipating short-term market changes, as other factors probably drive daily stock returns more significantly than sentiment ratings.

Implication: These results suggest that using sentiment analysis alone to forecast stock prices should be avoided by analysts and investors. Applied to longer-term analysis or in conjunction with other financial metrics, it might be more beneficial than using it for daily forecasts.
